```kotlin
%use lets-plot
%use dataframe
```

```kotlin
val df = DataFrame.readCSV("./data.csv")
df
```

```kotlin
val p = letsPlot(df.toMap()){
    x="n"
    y="height"
} + geomPoint(){color="frequency"} + geomLine(stat = Stat.smooth()){
    color="frequency"
} + facetGrid(y="frequency")
p + ggsize(800, 500)
```

```kotlin
val newDf = df.groupBy(){
    "frequency"<Int>()
}.aggregate { 
    val (one, two) = ("height"<Double>() / 100.0).toList().takeLast(2)
    (one) / (("n"<Int>()).max() - 1 ) into "tan"
    "n"<Int>().max() - 1 into "nMax"
    "height"<Double>().map{it / 100f} into "heights"
    "n"<Int>() into "ns"
}.split{"heights"<List<Double>>() and "ns"<List<Double>>()}.by { it.asIterable() }.intoRows().add("line"){
    "ns"<Int>()  * "tan"<Double>()
}
newDf
```

```kotlin
val pl = letsPlot(newDf.toMap()){
    x="ns"
    y="heights"
} + geomPoint(){color="frequency"} + geomLine(newDf.toMap()){
    x = "ns"
    y = "line"
}+ facetGrid(y="frequency")
pl + ggsize(800, 500)
```
